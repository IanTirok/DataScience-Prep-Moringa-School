```python
# Start by importing Numpy library
import numpy as np
```

```python
# Import Pandas library
import pandas as pd
```

```python
#Create dataframe from the linked Telcom_dataset csv file.
# Assign new names to the columns for consistency with similar dataframes in subsequent steps.
# Preview the first five rows
tcd_1 = pd.read_csv('Telcom_dataset.csv')
tcd_1.columns = ['PRODUCT','VALUE','DATETIME','CELL_ON_SITE','DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
tcd_1.loc[:5]
```

```python
# Create dataframe from the linked Telcom_dataset2 csv file.
# Change column names.
# Preview the dataframe
tcd_2 = pd.read_csv('Telcom_dataset2.csv')
tcd_2.columns = ['PRODUCT','VALUE','DATETIME','CELL_ON_SITE','DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
tcd_2.loc[:5]
```

```python
# Create dataframe from the linked Telcom_dataset3 csv file.
# Change column names.
# Preview the dataframe
tcd_3 = pd.read_csv('Telcom_dataset3.csv')
tcd_3.columns = ['PRODUCT','VALUE','DATETIME','CELL_ON_SITE','DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','CELL_ID','SITE_ID']
tcd_3.loc[:5]
```

```python
# Our next objective is to merge the three dataframes to have a common dataframe for the Telcom Datasets.
# Merge the three dataframes along their columns
# Assign the dataframe to variable 'tcd_c'
tcd_c = pd.concat([tcd_1,tcd_2,tcd_3], axis=0)
tcd_c
```

```python

# Drop duplicate records from our new dataframe
tcd_c.duplicated()
tcd_dup = tcd_c.drop_duplicates()
tcd_dup
```

```python

# Drop rows containing null records.
tcd_dup.isnull()
tcd_d = tcd_dup.dropna()
tcd_d
```

```python

# Dropping unnecessary columns from our dataframe
tcd_d = tcd_c.drop(columns=['DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B', 'CELL_ON_SITE', 'SITE_ID' ], axis = 0)
tcd_d
```

```python

# Create a dataframe for Cells Geo as 'cgd'.
# Create dataframe from the linked cells_geo.csv file.
# Preview the dataframe
cgd = pd.read_csv('cells_geo.csv', delimiter= ';')
cgd.loc[:5]
```

```python

# Drop duplicate records from our dataframe.
cgd.duplicated()
cgd_dup= cgd.drop_duplicates()
cgd_dup
```

```python

#Drop null records from our dataframe
cgd.isnull()
cgd_null = cgd.dropna()
cgd_null
```

```python
# Drop columns in cells_geo.csv
cgd_d = cgd_null.drop(columns=['Unnamed: 0', 'LOCALISATION', 'STATUS',	'DECOUPZONE',	'ZONENAME',	'LONGITUDE',	'LATITUDE', 'AREA', 'SITE_CODE'], axis = 0)
cgd_d
```

```python
# Describe the dataframe to get a view of the contents of the dataframe
cgd_d.describe()
```

```python
# Joining the telcom dataframe and cells geo dataframe on their common column cell id
fin_df = tcd_d.merge(cgd_d, on= ['CELL_ID'], how= 'outer')
fin_df
```

```python
#Drop null records from our dataframe
fin_df.isnull()
fin_df_null = fin_df.dropna()
fin_df_null
```

```python
# Drop duplicated values
fin_df.duplicated()
fin_df_dup= fin_df.drop_duplicates()
fin_df_dup
```

```python
# To find out at which duration of time the network was most used
fin_df_dup.groupby(['DATETIME'])['VALUE'].sum()
```

```python
#find out use of each product

fin_df_dup.groupby(['PRODUCT'])[['VALUE']].sum()
```
